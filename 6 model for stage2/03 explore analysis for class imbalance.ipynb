```python
# Data wrapper libraries
import pandas as pd
import numpy as np

# Visualization Libraries
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'retina'

from matplotlib.patches import Patch
from matplotlib.markers import MarkerStyle
import seaborn as sns
```

Let's take a casual look at the *variants* data.

```python
df_variants = pd.read_csv('../4 input for stage2/training_variants.csv').set_index('ID')
```

```python
plt.figure()
ax = df_variants['Gene'].value_counts().plot(kind='area')

ax.get_xaxis().set_ticks([])
ax.set_title('Gene Frequency Plot')
ax.set_xlabel('Gene')
ax.set_ylabel('Frequency')

plt.show()
```

And finally lets look at the class distribution.

```python
plt.figure(figsize=(12,8))
ax = df_variants['Class'].value_counts().plot(kind='bar')

ax.set_title('Class Distribution Over Entries', fontsize=22)
ax.set_xlabel('Class', fontsize=18)
ax.set_ylabel('Frequency', fontsize=18)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.show()
```
